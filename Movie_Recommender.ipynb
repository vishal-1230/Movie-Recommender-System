```python
import pandas as pd

movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")

movies.info()
credits.head()

movies = movies.merge(credits, on="title")

import ast

ast.literal_eval(credits['crew'][0])

movies = movies[["movie_id", "title", "genres", "keywords", "overview", "cast", "crew"]]
movies.head()
```

```python
import ast
movies["genres"][0]

def cleanGenres(genres):
  genres = ast.literal_eval(genres)
  new_genres = [i["name"] for i in genres]
  return " ".join(new_genres)

# print(cleanGenres('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'))
movies["genres"] = movies["genres"].apply(cleanGenres)
movies["keywords"] = movies["keywords"].apply(cleanGenres)
movies.head()
```

```python
movies["cast"][0]

def cleanCast(cast):
  cast = ast.literal_eval(cast)
  chars = ""
  for i in range(0, len(cast)):
    if i < 3:
      chars+=" "+cast[i]['character'].replace(" ", "")
    else:
      break
  return chars

def cleanCrew(crew):
  crew = ast.literal_eval(crew)
  for i in crew:
    if i['job'].lower() == "director":
      return i['name'].replace(" ", "")

movies['cast'] = movies['cast'].apply(cleanCast)
movies['crew'] = movies['crew'].apply(cleanCrew)
```

```python
movies.head()
```

```python
movies['tags'] = movies['genres'] + " " + movies["overview"] + " " + movies['keywords'] + " " + movies['cast'] + " " + movies['crew']
final_data = movies.drop(columns=["genres", "overview", "keywords", "cast", "crew"])
final_data.head()
```

```python
# final_data.head()

# print(final_data['tags'])

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
print(ps.stem("thriller"))
print(ps.stem("documentary"))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words="english")
# final_data['tags']
final_data['tags'] = final_data['tags'].fillna('')
# vector0 = cv.fit_transform(final_data['tags']).toarray()
# print(cv.get_feature_names_out()[0:100])

def stem(string):
  new_str = []
  for i in string.split():
    new_str.append(ps.stem(i))
  return " ".join(new_str)

print("Printing")
final_data['tags'] = final_data['tags'].apply(stem)
vector = cv.fit_transform(final_data['tags']).toarray()
print(cv.get_feature_names_out()[0:100])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

simi = cosine_similarity(vector)
```

```python
# sorted(simi[0], reverse=True)
sorted(list(enumerate(simi[0])), reverse=True, key=lambda x:x[1])
```

```python
def recommend(movie):
  movie_index = False
  for i in range(0, len(final_data)):
    if final_data['title'][i]==movie:
      movie_index = i
      print("Movie's index", i)
      break

  if movie_index:
    similarity_ids = sorted(list(enumerate(simi[movie_index])), reverse=True, key=lambda x:x[1])[1:10]
    return [final_data['title'][i[0]] for i in similarity_ids]
  else:
    return "Sorry, Movie not in database or there's a Typo"

recommend("Man of Steel")
```
